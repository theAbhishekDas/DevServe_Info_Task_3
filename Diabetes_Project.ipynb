**Split the data into training and testing data using the train_test_split function**

```python
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=7)
```

```python
#IMPORT LIBARIES for the model
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.metrics import roc_auc_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_recall_curve
```

```python
# Creating a dictionary of machine learning algorithms
models = {'Logistic Regression': LogisticRegression(),
          'K-nearest neighbors': KNeighborsClassifier(),
          'SVM': SVC(),
          'Decision Tree': DecisionTreeClassifier(),
          'Random Forest': RandomForestClassifier()}

# Loop through the dictionary and fit each algorithm to the training data
for name, model in models.items():
    model.fit(X_train, y_train)
    print(f"{name} trained successfully")
```

```python
# Looping through the dictionary then calculate the 
#accuracy score for each algorithm on the testing data
for name, model in models.items():
    y_pred = model.predict(X_test)
    acc = accuracy_score(y_test, y_pred)
    print(f"{name} accuracy score: {acc:.2f}")
```

```python
# Define the machine learning algorithm
model = DecisionTreeClassifier()
model.fit(X_train, y_train) # Fit the model to the training data

y_pred = model.predict(X_test) # Generate predictions on the testing data

acc = accuracy_score(y_test, y_pred) # Calculate the accuracy score
print(f"Accuracy score: {acc:.2f}")

cm = confusion_matrix(y_test, y_pred) # Calculate the confusion matrix
print("Confusion matrix:\n", cm)

cr = classification_report(y_test, y_pred)# Calculate the classification report
print("Classification report:\n", cr)
```
